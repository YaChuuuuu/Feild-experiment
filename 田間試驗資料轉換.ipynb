```python
import pandas as pd
wslist=['A','A','A','C','C','C','D','D','D']
columnlist=["編號","根冠1","根冠2","根冠3","根冠4","葉片數","中間葉長","中間葉寬","左邊葉長","左邊葉寬","葉面積(mm2)","第幾葉","炭疽級數","葉枯級數"]
for i in range(len(wslist)):
    #為了等下轉置，不讀取第一行做為表頭
    df_temp=pd.read_excel('raw_data/草莓田間數據-2021.8.16.xlsx',sheet_name=i,header=None)
    #轉置
    df_temp=df_temp.T
    #去掉多餘的資訊
    df_temp=df_temp.iloc[:,:14]
    #設定欄位
    df_temp.columns=columnlist
    #去除第一行
    df_temp=df_temp.drop(df_temp.index[0])
    
    #定義欄位格式
    for index,x in enumerate(columnlist):
        if index==5 or index==11:                    #當是葉片數和第幾葉時，要轉成整數
            for n in range(len(df_temp[x])):
                if df_temp[x].notnull().iloc[n]==True:
                    df_temp.iloc[n,index]=int(df_temp.iloc[n,index])
        elif index==6 or index==7 or index==8 or index==9 or index==10:    #當是葉長、葉寬相關資訊時，要四捨五入數
            for n in range(len(df_temp[x])):
                if df_temp[x].notnull().iloc[n]==True:
                    df_temp.iloc[n,index]=round(df_temp.iloc[n,index],2)
        elif index==0 or index==12 or index==13:    #編號、病害級數要轉整數
            df_temp[x]=df_temp[x].fillna(value=0)
            df_temp[x]=df_temp[x].astype("int16")
    
    #編號前加上處理的場域代號
    for n,x in enumerate(list(df_temp["編號"])):
        df_temp.iloc[n,0]=wslist[i]+str(x)
    
    #加入月份欄位
    month=[]
    for n in range(len(df_temp["編號"])):
        month.append('Aug')
    df_temp.insert(14,column="時間",value=month)
    if i ==0:
        df_Aug=df_temp.copy()
    else:
        df_Aug=df_Aug.append(df_temp)
len(df_Aug)
```
